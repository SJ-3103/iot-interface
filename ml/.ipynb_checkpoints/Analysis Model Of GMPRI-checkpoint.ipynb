# Train Test Splitting

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(plant,test_size=0.2,random_state=42)
print(f"Rows in train set:{len(train_set)}\nRows in test set:{len(test_set)}")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(plant,plant['Temperature']):
    strat_train_set=plant.loc[train_index]
    strat_test_set=plant.loc[test_index]
```

```python
strat_test_set['Temperature'].value_counts()
```

```python
strat_train_set['Temperature'].value_counts()
```

```python
plant=strat_train_set.copy()
```

```python
plant=strat_train_set.drop("Humidity",axis=1)
plant_labels=strat_train_set["Humidity"].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(plant)
```

```python
imputer.statistics_
```

```python
X=imputer.transform(plant)
```

```python
plant_tr=pd.DataFrame(X,columns=plant.columns)
```

```python
plant_tr.describe()
```

# Creating Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler()),
])
```

```python
plant_num_tr=my_pipeline.fit_transform(plant_tr)
```

```python
plant_num_tr.shape
```

# Selecting a desired model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
#model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(plant_num_tr,plant_labels)
```

```python
some_data=plant.iloc[:5]
```

```python
some_labels=plant_labels.iloc[:5]
```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

# Evaluating the model

```python
from sklearn.metrics import mean_squared_error
plant_predictions=model.predict(plant_num_tr)
mse=mean_squared_error(plant_labels,plant_predictions)
rmse=np.sqrt(mse)
rmse
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,plant_num_tr,plant_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores:",scores)
    print("Mean:",scores.mean())
    print("Standard deviation:",scores.std())
print_scores(rmse_scores)
```

# Saving the model

```python
from joblib import dump,load
dump(model,'Price_Predictor.joblib')
```

# Testing The model

```python
X_test=strat_test_set.drop("Humidity",axis=1)
Y_test=strat_test_set["Humidity"].copy()
X_test_prepared=my_pipeline.transform(X_test)
final_predictions=model.predict(X_test_prepared)
final_mse=mean_squared_error(Y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
final_rmse
```
